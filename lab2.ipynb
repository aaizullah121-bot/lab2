```python
import matplotlib.pyplot as plt

# Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [15, 30, 45, 10]

# Create bar graph
plt.bar(categories, values)

# Add labels and title
plt.xlabel('Fruits')
plt.ylabel('Quantity Sold')
plt.title('Fruit Sales Data')

# Display the plot
plt.show()
```

```python
# 1. Define a list of 10 items
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry", "Fig", "Grape", "Honeydew", "Kiwi", "Lemon"]
print(f"Original List (Length {len(fruits)}): {fruits}")

# 2. Add an item
fruits.append("Mango")
print(f"After Append: {fruits}")

# 3. Remove an item (removes 'Cherry')
fruits.remove("Cherry")
print(f"After Removal: {fruits}")

# 4. Final count
print(f"Final Item Count: {len(fruits)}")
```

```python
# 1. Create a tuple of 10 fruits
fruits = ("Apple", "Banana", "Cherry", "Date", "Elderberry", "Fig", "Grape", "Honeydew", "Kiwi", "Lemon")

# 2. Print the 1st fruit (Index 0)
print(f"First fruit: {fruits[0]}")

# 3. Print the last fruit (Index -1)
print(f"Last fruit: {fruits[-1]}")
```

```python
# 1. Setup
set_a = {10, 20, 30, 40}
set_b = {30, 40, 50, 60}

# 2. Operations
union_set = set_a | set_b        # All items
intersect_set = set_a & set_b    # Common items
diff_set = set_a - set_b         # Only in A

# 3. Print Results
print(f"Union: {union_set}")
print(f"Intersection: {intersect_set}")
print(f"Difference (A-B): {diff_set}")
```

```python
# 1. Create a dictionary
car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

# 2. Print Keys and Values
print("Keys:", car.keys())
print("Values:", car.values())

# 3. Update a Value
car["year"] = 2024
# Alternative way to update
car.update({"model": "Mach-E"})
print("\nAfter Update:", car)

# 4. Delete a key-value pair
# Method 1: Using pop()
car.pop("brand")

# Method 2: Using del
# del car["model"]

print("After Deletion:", car)
```
