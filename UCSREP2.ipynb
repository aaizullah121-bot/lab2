**UCS GRAPH IMPLEMENTATION**

```python
import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('E', 3), ('D', 1)],
    'C': [('F', 5)],
    'D': [('G', 2)],
    'E': [('D', 1)],
    'F': [('G', 2)],
    'G': []
}

def ucs(graph, start, goal):
    pq = [(0, start, [])]  # (cost, node, path)
    visited = set()

    while pq:
        cost, node, path = heapq.heappop(pq)

        if node in visited:
            continue

        visited.add(node)
        path = path + [node]

        if node == goal:
            return path, cost

        for neighbor, weight in graph[node]:
            heapq.heappush(pq, (cost + weight, neighbor, path))

    return None, float("inf")

path, cost = ucs(graph, 'A', 'G')

print("Path:", path)
print("Cost:", cost)
```
